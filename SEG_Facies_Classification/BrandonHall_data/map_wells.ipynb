```python
#plot locations of kansas wells where cores were id'd
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

from pandas import set_option

plt.close('all')

set_option("display.max_rows", 12)

well_data = pd.read_csv('KSdata/training_wells.csv')
test_data = pd.read_csv('KSdata/test_wells.csv')

plt.figure()
map = Basemap(llcrnrlon=-102,llcrnrlat=36.5,urcrnrlon=-100.5,urcrnrlat=38.5,
            projection='lcc',lat_1=33,lat_2=45,lon_0=-95, resolution='h')

map.drawcounties()
map.drawstates(color='black', linewidth=1)
map.shadedrelief()

x, y = map(well_data['LONGITUDE'].values,
           well_data['LATITUDE'].values,)
s1 = map.scatter(x, y, alpha=0.7, edgecolors='none')
x, y = map(test_data['LONGITUDE'].values,
           test_data['LATITUDE'].values,)
s2 = map.scatter(x, y, alpha=0.7, edgecolors='none', c='r')
plt.legend([s1,s2], ['Training Wells', 'Blind Test Wells'])
plt.title('Location of Study Wells')

plt.show()
```
